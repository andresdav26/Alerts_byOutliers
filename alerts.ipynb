```python
df['Outlier reprocesos general']= df['Reprocesos estados'].median()+df['Reprocesos estados'].std()
df['Outlier reprocesos combinaciones general']= df['Reprocesos Combinaciones estados'].median()+df['Reprocesos Combinaciones estados'].std()
```

```python
state_out_reprocess=[]
list_states_reprocess=df['Estado'].drop_duplicates().to_list()
for i,l in enumerate(list_states_reprocess):
    out_day_sta_reprocess=np.percentile(df[df['Estado']==l]['Reprocesos estados'],65)
    state_out_reprocess1={
        'Estado':l,
        'Outlier reprocesos estados':out_day_sta_reprocess
    }
    state_out_reprocess.append(state_out_reprocess1)
```

```python
state_out_reprocess=pd.DataFrame(state_out_reprocess)
df=df=pd.merge(left=df,right=state_out_reprocess)
df['Alerta reprocesos estados '] = df['Reprocesos estados']>df['Outlier reprocesos estados']
```

```python
state_out_reprocess_comb=[]
list_states_reprocess_comb=df['Combinacion estado'].drop_duplicates().to_list()
for i,l in enumerate(list_states_reprocess_comb):
    out_day_sta_reprocess_comb=np.percentile(df[df['Combinacion estado']==l]['Reprocesos Combinaciones estados'],65)
    state_out_reprocess1_comb={
        'Combinacion estado':l,
        'Outlier reprocesos combinaciones':out_day_sta_reprocess_comb
    }
    state_out_reprocess_comb.append(state_out_reprocess1_comb)
```

```python
state_out_reprocess_comb=pd.DataFrame(state_out_reprocess_comb)
df=df=pd.merge(left=df,right=state_out_reprocess_comb)
df['Alerta reprocesos combinaciones estados'] = df['Reprocesos Combinaciones estados']>df['Outlier reprocesos combinaciones']
```

**Dias laborados por combinaciones outliers:**

```python
state_out_comb=[]
list_states_comb=df['Combinacion estado'].drop_duplicates().to_list()
for i,l in enumerate(list_states_comb):
    q3_day_sta_comb=df[df['Combinacion estado']==l]['Dias laborados por estado'].quantile(0.75)
    q1_day_sta_comb=df[df['Combinacion estado']==l]['Dias laborados por estado'].quantile(0.25)
    IQR_days_sta_comb=q3_day_sta_comb-q1_day_sta_comb
    out_day_sta_comb=q3_day_sta_comb+ 1.5*IQR_days_sta_comb
    state_out_comb1={
        'Combinacion estado':l,
        'Outlier dias combinaciones' :out_day_sta_comb
    }
    state_out_comb.append(state_out_comb1)
```

**Documentación columnas**

Cantidad reprocesos por estado: Registra la cantidad de veces que pasa un estado por radicado.
Cantidad reprocesos por combinaciones: Registra la cantidad de veces que pasa una combinación por radicado.
Outlier_reprocesos: Es el valor limite , los estados que tienen un número superior a este son valores atípicos de reprocesos por estado.
Outlier_reprocesos_comb: Es el valor limite , las combinaciones que tienen un número superior a este son valores atípicos de reprocesos por combinaciones.
Outlier dias combinaciones: Es la cantidad de dias limite que normalmente tiene cada proceso.
Alerta dias por combinaciones: Estos es una descripcion de que combinaciones superan los dias de outlier.
Outlier dias estados: Es la cantidad de dias limite que normalmente tiene cada combinación.
Alerta dias por estados: Estos es una descripcion de que estados superan los dias de outlier.
Alerta dias por radicado: Radicados que superan la cantidad de dias que normalmente tienen.
Dias laborados por radicado: Son los dias laborados por radicado.
Radicados con dias superiores a ANS: Son los radicados que sus dias superan los dias normalmente marcados de ANS.
Outlier reprocesos estados: Indica cuáles son los outliers de cada estado individual.
Outlier reprocesos combinaciones: Indica cuáles son los outliers de cada combinación individial.

```python
import sklearn
from sklearn import tree
from sklearn.model_selection import KFold
```

**Predicción alerta de dias laborados por radicado:**

```python
predict_df1=predict_df.drop(columns=[ 'Dias laborados por estado','Estado', 'Combinacion estado','Alerta reprocesos estados ',
       'Alerta reprocesos combinaciones estados',
       'Alerta dias por combinaciones', 'Alerta dias por estados', 'Radicados con dias superiores a ANS','Valor'])
predict_df1=predict_df1.drop_duplicates(subset=['Radicado'])
list_index=[]
for u in range(len(predict_df1)):
    list_index.append(u)
predict_df1['index']=list_index
predict_df1=predict_df1.set_index('index')
cv = KFold(n_splits=10) # Numero deseado de "folds" que haremos
accuracies = list()
max_attributes = len(list(predict_df1))
depth_range = range(1, max_attributes + 1)

# Testearemos la profundidad de 1 a cantidad de atributos +1
for depth in depth_range:
    fold_accuracy = []
    tree_model = tree.DecisionTreeClassifier(criterion='entropy',
                                             min_samples_split=20,
                                             min_samples_leaf=5,
                                             max_depth = depth,
                                             class_weight={1:51.38})
    for train_fold, valid_fold in cv.split(predict_df1):
        f_train = predict_df1.loc[train_fold] 
        f_valid = predict_df1.loc[valid_fold] 

        model = tree_model.fit(X = f_train.drop(['Alerta dias por radicado'], axis=1), 
                               y = f_train['Alerta dias por radicado']) 
        valid_acc = model.score(X = f_valid.drop(['Alerta dias por radicado'], axis=1), 
                                y = f_valid['Alerta dias por radicado']) # calculamos la precision con el segmento de validacion
        fold_accuracy.append(valid_acc)

    avg = sum(fold_accuracy)/len(fold_accuracy)
    accuracies.append(avg)
    
# Mostramos los resultados obtenidos
prub = pd.DataFrame({"Max Depth": depth_range, "Average Accuracy": accuracies})
prub = prub[["Max Depth", "Average Accuracy"]]
print(prub.to_string(index=False))
```

```python
y_train = predict_df1['Alerta dias por radicado']
x_train = predict_df1.drop(['Alerta dias por radicado'], axis=1).values 
 
# Crear Arbol de decision con profundidad = 4
decision_tree = tree.DecisionTreeClassifier(criterion='entropy',
                                            min_samples_split=20,
                                            min_samples_leaf=5,
                                            max_depth = 8,
                                            class_weight={1:51.38})
decision_tree.fit(x_train, y_train)
acc_decision_tree = round(decision_tree.score(x_train, y_train) * 100, 2)
print(acc_decision_tree)
```
